```python
import matplotlib.pyplot as plt 
theta1=0
theta2=0
derv_theta=(2*theta1)+(2*theta2)-4
itr=1
lr=0.1
t1=[]
t1.append(theta1)
t2=[]
t2.append(theta2)
J=[]
j=pow((theta1+theta2-2),2)
J.append(j)
while itr!=100:
    theta1=theta1-(lr*derv_theta)
    theta2=theta2-(lr*derv_theta)
    j=pow((theta1+theta2-2),2)
    derv_theta=(2*theta1)+(2*theta2)-4
    t1.append(theta1)
    t2.append(theta2)
    J.append(j)
    itr+=1

ax = plt.axes(projection='3d')
ax.plot3D(t1,t2,J, 'gray')
ax.scatter(theta1,theta2,j)    
```
